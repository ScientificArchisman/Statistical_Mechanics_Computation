```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as const
plt.style.use(["science", "notebook", "grid"])
```

$$
u_\nu d\nu = \frac{8 \pi \nu^2}{c^3} \frac{h\nu}{e^{\frac{h\nu}{kT}}-1} d\nu

$$


```python
def planck(f, T):
    """Planck function for a given wavelength and temperature"""
    return (8 * np.pi * const.h * f**3 / (const.c**2 * (np.exp(const.h * f / (const.k * T)) - 1)))

```

```python
f = np.linspace(10**12, 150*10**12, 1000)
T = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 10000])
plt.figure(figsize = (15, 8))

for temp in T:
    plt.plot(f, planck(f, temp), label = str(temp) + " K")

plt.legend()
```
