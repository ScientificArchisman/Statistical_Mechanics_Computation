# Random Walk in 1D
---
### Steps to follow:
1. Import the required libraries
2. Define the number of steps
3. Define the number of random walks
4. Define the initial position
5. Define the step size
6. Define the step direction
7. Define the random walk
8. Plot the random walk
9. Plot the histogram of the random walk
10. Plot the cumulative sum of the random walk
---

````python
"""Capitalised variables are constants"""

N_STEPS = 1000 # Number of steps
N_WALKS = 200 # Number of random walks
x0 = 0 # Initial position
STEP_SIZE = 1  # Step size
````

### At first we define a function to generate a random walk in 1D

````python
def random_walk_1D(n_steps:int, n_walks:int, x0:int, step_size:float) -> np.ndarray:
    """Returns a 2D array of random walks"""
    # Create array of random steps
    steps = np.random.choice([-step_size, step_size], size=(n_walks, n_steps))
    # Create array of random walks
    walks = np.cumsum(steps, axis=1)
    # Add initial position to each walk
    walks  = np.insert(walks, 0, x0, axis=1)
    return walks # Return the wals
````

### Then we define a function to plot the random walk

````python
def plot_random_walks(walks:np.ndarray, title:str = "Random walk in 1D") -> plt.Axes:
    """Plots the random walks"""
    plt.plot(walks.T, alpha=0.5)
    
    plt.title(title)
    plt.xlabel("Step")
    plt.ylabel("Position")
    plt.show()
````

````python
plt.figure(figsize=(10, 6))
random_walks = random_walk_1D(N_STEPS, N_WALKS, x0, STEP_SIZE)
plot_random_walks(random_walks)
````
